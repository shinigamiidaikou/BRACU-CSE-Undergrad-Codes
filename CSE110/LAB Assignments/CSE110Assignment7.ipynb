### Task 1

Suppose you have a list named my_list as given below.

<font face=consolas>my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]</font>

Now use bubble sort to sort my_list into ascending order.

===================================================================

**Output:**\
<font face=consolas>[1, 2, 2, 3, 5, 6, 10, 11, 12, 14, 15, 17, 18, 20, 29]</font>

```python
#Task 1 (Bubble Sort)

my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]

for i in range(len(my_list)-1):
    for j in range(len(my_list)-i-1):
        if my_list[j] > my_list[j+1]:
            temp = my_list[j+1]
            my_list[j+1] = my_list[j]
            my_list[j] = temp

print(my_list)
```

### Task 4

Suppose you have a sitting arrangement of the students who will give the final exam in a list. The list contains the last two digits of their student ID as given below.

<font face=consolas>sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]</font>

Now you want to organize the sitting arrangement of the students in your own way. You decide to sort all the students in the even indices of the list in ascending order and all the students in the odd indices of the list in descending order.\
So, write a python program that organizes the list for you in this way.

===================================================================

**Output:**\
<font face=consolas>[10, 100, 11, 80, 12, 70, 20, 56, 22, 30, 58, 16, 70, 15]</font>


```python
#Task 4

sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]
even_indices = list()
odd_indices = list()

# Seperation:
for index in range(len(sitting_list)):
    if index % 2 == 0:
        even_indices.append(sitting_list[index])
    else:
        odd_indices.append(sitting_list[index])

# Sorting Even Indices (Ascending Order):
for i in range(len(even_indices) - 1):
    for j in range(len(even_indices) - i - 1):
        if even_indices[j] > even_indices[j + 1]:
            temp = even_indices[j + 1]
            even_indices[j + 1] = even_indices[j]
            even_indices[j] = temp

# Sorting Odd Indices (Descending Order):
for i in range(len(odd_indices)-1):
    maximum = odd_indices[i]
    max_index = i
    for j in range(i+1, len(odd_indices)):
        if odd_indices[j] > maximum:
            maximum = odd_indices[j]
            max_index = j
    temp = odd_indices[i]
    odd_indices[i] = odd_indices[max_index]
    odd_indices[max_index] = temp

# Creating Desired Sorted Sitting List:
if len(sitting_list) % 2 == 0: # Even Length
    sorted_sitting_list = list()
    for i in range(len(even_indices)):
        sorted_sitting_list.append(even_indices[i])
        sorted_sitting_list.append(odd_indices[i])
else:
    sorted_sitting_list = list() # Odd Length
    for i in range(len(even_indices) - 1):
        sorted_sitting_list.append(even_indices[i])
        sorted_sitting_list.append(odd_indices[i])
    sorted_sitting_list.append(even_indices[-1])

print(sorted_sitting_list)
```

### Task 7
Write a python program that takes two lists from the user, merges the two lists, sorts the resulting list, and then finds the median of the elements in the two lists.

===================================================================

**Sample Input 1**

<font face=consolas>
list_one = [1, 2, 1, 4]<br>
list_two = [5, 4, 1]</font>

**Sample Output 1**

<font face=consolas>
Sorted list = [1, 1, 1, 2, 4, 4, 5]<br>
Median = 2</font>

===================================================================

**Sample Input 2**

<font face=consolas>
list_one = [1, 7, 9, 10]<br>
list_two = [2, 7, 6, 5]</font>

**Sample Output 2**

<font face=consolas>
Sorted list = [1, 2, 5, 6, 7, 7, 9, 10]<br>
Median = 6.5</font>


```python
# Task 7

firstList_in = input("list_one = ")
secondList_in = input("list_two = ")
merged_input = firstList_in[1:-1] + ", " + secondList_in[1:-1]
merged_input = merged_input.replace(" ", "")
mergedList = list()

# Creating Merged List
temp = str()
for i in range(len(merged_input)):
    if merged_input[i] != ",":
        temp += merged_input[i]
        if i == len(merged_input) - 1:
            mergedList.append(int(temp))
            break
    else:
        mergedList.append(int(temp))
        temp = str()

# Sort (Ascending):
for i in range(len(mergedList) - 1):
    for j in range(len(mergedList) - i - 1):
        if mergedList[j] > mergedList[j + 1]:
            temp = mergedList[j + 1]
            mergedList[j + 1] = mergedList[j]
            mergedList[j] = temp

# Finding Median
if len(mergedList) % 2 == 0:
    median = ( mergedList[(len(mergedList)//2)-1] + mergedList[len(mergedList)//2] ) / 2
    if median == int(median):
        median = int(median)
else:
    median = mergedList[(len(mergedList)-1)//2]

print("\nSorted List =", mergedList)
print("Median =", median)
```

### Task 9

===================================================================

**Hint:**

The formula of distance = $\sqrt{(\Delta x)^2+(\Delta y)^2}$\
As you are calculating the distance from the origin $(0,0)$, you can simply use

distance = $\sqrt{x^2+y^2}$

You can create a list of distances from each point and sort that list using your personal favorite sorting algorithm.

===================================================================

**Sample Input 1**

<font face=consolas>
points = [(5,3), (2,9), (-2,7), (-3,-4), (0,6), (7,-2)]
</font>

```python
# Task 9

import math

points = [(5,3), (2,9), (-2,7), (-3,-4), (0,6), (7,-2), (3,4), (4,3)]
distance_list = list()

# Creating List of Distances
for tpl in points:
    distance = math.sqrt(tpl[0]**2 + tpl[1]**2)
    distance_list.append(distance)

# Finding Minimum Value of Distance
minimum = distance_list[0]
for i in range(len(distance_list)):
    if minimum >= distance_list[i]:
        minimum = distance_list[i]
print(f"Minimum distance = {minimum}")

# Checking tuples
for tpl in points:
    if math.sqrt(tpl[0]**2 + tpl[1]**2) == minimum:
        print(f"Here the closest point is {tpl} which has a distance of {minimum} from the origin.")
        # print within loop & condition for finding all possible same results
```
