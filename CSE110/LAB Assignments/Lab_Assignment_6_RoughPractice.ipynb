```python
# Task 2
def fibonacci(Limit):
    number1 = 0
    number2 = 1
    print(number1, end=" ")
    
    while number2 <= Limit:
        print(number2, end = " ")
        temporary = number1
        number1 = number2
        number2 = temporary + number2

fibonacci(8)
```

```python
# Task 7

def show_palindrome(num):
    
    string = str()

    for i in range(1, num+1):
        string += str(i)
    for i in range(num-1, 0, -1):
        string += str(i)
    
    return string

print(show_palindrome(9))
print(show_palindrome(5))
```

```python
# Task 11

def rem_duplicate(tupl):
    
    List = []
    
    for i in tupl:
        if i not in List:
            List.append(i)
    
    tupl = tuple(List)
    
    return tupl

print(rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0)))
print(rem_duplicate(("Hi", "Hello", 1, 0, 0, 3, 3, "Hello",'b', 'b', [2,4])))
```

```python
# Task 12

def max_Listing(List):
    
    new_List = list()
    rem_count = 0
    
    for elmnt in List:
        
        if List.count(elmnt) == 1 and elmnt not in new_List:
            new_List.append(elmnt)
        elif List.count(elmnt) == 2 and elmnt not in new_List:
            new_List.append(elmnt)
        elif List.count(elmnt) == 2 and elmnt in new_List:
            if new_List.count(elmnt) == 1:
                new_List.append(elmnt)
        elif List.count(elmnt) > 2 and elmnt not in new_List:
            new_List.append(elmnt)
        elif List.count(elmnt) > 2 and elmnt in new_List:
            if new_List.count(elmnt) == 1:
                new_List.append(elmnt)
                rem_count += List.count(elmnt) - 2
    
    print("Removed:", rem_count)
    return new_List

print(max_Listing([1, 1, 4, 5, 2, 2, 2, 3, 3, 3, 3, 4, 5, 6, 8, 8]))
print(max_Listing([10, 15, 10, 15, 20]))
```

```python
# Task 12 (efficient version)

def max_Listing(List):
	new_List = list()
	rem_count = 0
	for elmnt in List:
		if elmnt not in new_List:
			if List.count(elmnt) == 1:
				new_List.append(elmnt)
			elif List.count(elmnt) == 2:
				new_List += [elmnt, elmnt]
			elif List.count(elmnt) > 2:
				new_List += [elmnt, elmnt]
				rem_count += List.count(elmnt) - 2
	
	print("Removed:", rem_count)
	return new_List

print(max_Listing([1, 1, 4, 5, 2, 2, 2, 3, 3, 3, 3, 4, 5, 6, 8, 8]))
print(max_Listing([10, 15, 10, 15, 20]))
```

```python
# Task 13

def basic_calc(operator_sign, first_operand, second_operand):
    
    if operator_sign == '"+"':
        result = first_operand + second_operand
    elif operator_sign == '"-"':
        result = first_operand - second_operand
    elif operator_sign == '"/"':
        result = first_operand / second_operand
    elif operator_sign == '"*"':
        result = first_operand * second_operand
    
    return result

sign = input()
first_num = int(input())
second_num = int(input())

print(basic_calc(sign, first_num, second_num))
```

```python
# Task 14

sentence = input()
position = int(input())

def string_divider(sentence, position):
    sentence = sentence[1:-1]
    new_sentence = sentence[0]
    removed_chars = str()

    for i in range(1, len(sentence)):
        if i % position != 0:
            new_sentence += sentence[i]
        else:
            removed_chars += sentence[i]
    
    new_sentence += removed_chars
    return new_sentence

print(string_divider(sentence, position))
```

```python
# Task 23

amount = int(input())

def splitting_money(amount):
    
    note_list = [500, 100, 50, 20, 10, 5, 2, 1]
    string = str()

    for note in note_list:
        if amount // note != 0:
            note_quantity = amount // note
            if amount - (note_quantity * note) == 0:
                amount = 0
                string += str(note) + " Taka: " + str(note_quantity) + " note(s)"
            else:
                amount -= note_quantity * note
                string += str(note) + " Taka: " + str(note_quantity) + " note(s)\n"
    
    return string

print(splitting_money(amount))
```

```python
# Task 25

start_val_inc = int(input())
end_val_exc = int(input())
first_div = int(input())
second_div = int(input())

def divisable_summ(start_val_inc, end_val_exc, first_div, second_div):

    summ = int()

    for number in range(start_val_inc, end_val_exc):
        if not(number % first_div == 0 and number % second_div == 0):
            if number % first_div == 0 or number % second_div == 0:
                summ += number
    
    return summ

print(divisable_summ(start_val_inc, end_val_exc, first_div, second_div))
```
