```python
# Task 11

string = input()
char_freq = {}
char_List = []

for char in string[1:-1]:
    if char != " ":
        if 97 <= ord(char) <= 122 and char not in char_List:
            char_List.append(char)
        elif 65 <= ord(char) <= 90 and chr(ord(char) + 32) not in char_List:
            char_List.append(chr(ord(char) + 32))

for item in char_List:
    counter = 0
    for char in string[1:-1]:
        if char == item or chr(ord(char) + 32) == item:
            counter += 1
    char_freq [item] = counter

print(char_freq)
```

<font color='pale blue'>[Empty items means keys without any values (None)].</font>\

### **Task 24**
---
Given a list of tuples, your task is to group the tuples based on the second element in the tuples as shown in the examples below. We can achieve this using dictionary by checking the second element in each tuple.\
[You do not need to take tuple as input and can assume that it is given as below]\
\===================================================================\
***Example 1:***\
**Given**:\
[(20, 80), (31, 80), (1, 22), (88, 11), (27, 11)]\
**Output:**\
{80: [(20, 80), (31, 80)],\
11: [(88, 11), (27, 11)],\
22: [(1, 22)]}\
\===================================================================\
***Example 2:***\
**Given:**\
[(20, 'Sad'), (31, 'Sad'), (88, 'NotSad'), (27, 'NotSad')]\
**Output:**\
{'NotSad': [(88, 'NotSad'), (27, 'NotSad')],\
'Sad: [(20, 'Sad'), (31, 'Sad')]}\
\===================================================================


```python
given_tuple = [(20, 80), (31, 80), (1, 22), (88, 11), (27, 11)]
category_dict = dict()

for item in given_tuple:
    category_dict [item[1]] = []

for item in given_tuple:
    if item[1] in category_dict.keys():
        category_dict[item[1]].append(item)

print(category_dict)
```
