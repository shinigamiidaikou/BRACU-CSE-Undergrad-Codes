### My Question
**Write a Python program that takes the value of n from the user and calculates the value of y using the following equation.**\
$y = 3^2 + 6^2 - 9^2 + 12^2 + 15^2 - 18^2 + .......... (3n)^2$ \

[Here, ' ^ ' symbol represents exponentiation. You can import the 'math' module in this problem if necessary.]
### =============================================================
**Sample Input 1:**\
4\
**Sample Output 1:**\
Value of y is: 108

**Explanation:**\
For n = 4, if you implement the equation accurately in your code, then you will get y = 108 like the sample output.  
### =============================================================
**Sample Input 2:**\
9\
**Sample Output 2:**\
Value of y is: 297

**Explanation:**\
For n = 9, if you implement the equation accurately in your code, then you will get y = 297 like the sample output. 
### =============================================================

```python
# My Question Answer

Limit = int(input())
summ = 0

for counter in range(1, Limit + 1):
    if counter % 3 == 0:
        summ += -((3 * counter) ** 2)
    else:
        summ += ((3 * counter) ** 2)

print(f"Value of y is: {summ}")
```

```python
num = int(input())
temp = num
count = 0

while temp != 0:
    temp = temp // 10
    count += 1

if num % 2 == 0:
    summ = 0
    for i in range(1, num + 1):
        if i % 3 == 0:
            summ = summ + i
    product = summ * count
else:
    summ = 0
    for i in range(1, num + 1):
        if i % 5 == 0:
            summ = summ + i
    product = summ * count

print("Result =",product)
```
