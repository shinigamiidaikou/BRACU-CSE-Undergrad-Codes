### Code 1:

```python
def product_of_list(dictionary):
	list1 = list()
	for value in dictionary.values():
		value1 = ord(value[0])
		value2 = ord(value[1])
		list1.append(value1 * value2)
	return tuple(list1)


print(product_of_list({"A": ["a", "m"]}))
```
