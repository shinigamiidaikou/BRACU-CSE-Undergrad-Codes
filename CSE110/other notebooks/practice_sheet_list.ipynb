### Problem 1:

Assume, you have been given two lists. Print both lists simultaneously such that list1 should display items in original order and list2 in reverse order

=====================================================

**Given lists:**\
List1 = [10,20,30,40]\
List2 = [100,200,300,400]

**Sample Output:**\
10 400\
20 30\
30 200\
40 100

=====================================================

**Given lists:**\
List1 = [90,30,-30,20,12]\
List2 = [-8,-5,-2,100,6]

**Sample Output:**\
90 6\
30 100\
-30 -2\
20 -5\
12 -8

```python
def custom_print(list_one, list_two):
	
	for i in range(len(list_one)):
		print(f"{list_one[i]} {list_two[-i-1]}")

List1 = [10,20,30,40]
List2 = [100,200,300,400]
custom_print(List1, List2)
print()

List1 = [90,30,-30,20,12]
List2 = [-8,-5,-2,100,6]
custom_print(List1, List2)
```

### Problem 2:

Write a Python program to reverse strings in a given list of string values without using reverse() or reversed().

==============================================================

**Sample Input:** ['Red','Green','Blue','White','Black']

**Sample Output:** ['deR', 'neerG', 'eulB', 'etihW', 'kcalB']

```python
def list_item_reverser(List):

	for i in range(len(List)):
		new_string = str()
		for j in range(len(List[i])):
			new_string += List[i][-j-1]
		List[i] = new_string
	
	print(List)


list_input = input()
string_list = list()
temp = str()
for i in range(len(list_input[1:-1])):
	if list_input[1:-1][i] != ",":
		if list_input[1:-1][i] != "'":
			temp += list_input[1:-1][i]
			if i == len(list_input[1:-1]) - 2:
				string_list.append(temp)
				break
	else:
		string_list.append(temp)
		temp = str()


print(string_list)

list_item_reverser(string_list)
```

### Problem 3:

Write a Python program that takes a list from the user. Then find the first even and first odd number in the list.

=====================================================

**Sample Input1:**\
[1, 3, 5, 7, 4, 1, 6, 8]

**Sample Output1:**\
First even: 4 and First odd: 1

=====================================================

**Sample Input2:**\
[55, 5, 7, 9, 11]

**Sample Output2:**\
First even: not found and First odd: 55

=====================================================

**Sample Input3:**\
[50, 52, 6, 2, 10]

**Sample Output3:**\
First even: 50 and First odd: not found

```python
list_input = input()

num_list = list_input[1:-1].split(", ")
for i in range(len(num_list)):
	num_list[i] = int(num_list[i])

first_even = "not found"
for i in range(len(num_list)):
	if num_list[i] % 2 == 0:
		first_even = num_list[i]
		break

first_odd = "not found"
for i in range(len(num_list)):
	if num_list[i] % 2 != 0:
		first_odd = num_list[i]
		break

print("First even:", first_even, 'and First odd:', first_odd)
```

### Problem 4:

Assume, you have been given a list T and a number, N. Now, create a list by concatenating the input list which range goes from 1 to N.

=====================================================

**Sample input1:**\
T = ['p', 'q']\
N = 5

**Sample Output1:**\
['p1', 'q1', 'p2', 'q2', 'p3', 'q3', 'p4', 'q4', 'p5', 'q5']

=====================================================

**Sample input2:**\
T = ['p', 'q', 'r']\
N = 3

**Sample Output2:**\
['p1', 'q1', 'r1', 'p2', 'q2', 'r2', 'p3', 'q3', 'r3']

=====================================================

**Sample input3:**\
T = ['p', 'q', 'r', 's']\
N = 2

**Sample Output3:**\
['p1', 'q1', 'r1', 's1', 'p2', 'q2', 'r2', 's2']

```python
List = input()
string_list = list()
temp = str()
for i in range(len(List[1:-1])):
	if List[1:-1][i] != ",":
		if List[1:-1][i] not in "' ":
			temp += List[1:-1][i]
			if i == len(List[1:-1]) - 2:
				string_list.append(temp)
				break
	else:
		string_list.append(temp)
		temp = str()

limit = int(input())

concatenated_rep_list = list()
for i in range(1, limit + 1):
	for j in range(len(string_list)):
		concatenated_rep_list.append(string_list[j] + str(i))


print(concatenated_rep_list)
```
