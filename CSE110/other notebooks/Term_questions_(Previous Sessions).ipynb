### Question-1:
 
Write a Python program that reads a number and finds the sum of the series\
$1 + 33 + 111 + 3333 + .... + N$ terms. 

Then it puts the reverse digit of the sum to a list and prints it.\
<font color=red>**[Cannot use built-in reverse() or reversed()]**</font>

=======================================================================

**Sample Input 1:**

5

**Sample Output 1:**

Sum of series: 14589\
Reverse sum of series in list: [9, 8, 5, 4, 1]

**Explanation 1 :**\
The input is 5 so the series will go on till 5th term. So the series becomes, 1 +33 + 111 + 3333 + 11111.\
So the sum of the series becomes 14589.\
The reverse of the sum is 98541.\
So te reverse digits are 9, 8, 5, 4\
And Thus putting them into a list, the 
output becomes [9, 8, 5, 4, 1] 

=======================================================================

**Sample Input 2:**

3

**Sample Output 2:**

Sum of series: 145\
Reverse sum of series in list: [5, 4, 1]

**Explanation 2:**\
The input is 3 so the series will go on till 3rd term. So the series becomes, 1 + 33 + 111.\
So the sum of the series is 145.\
The reverse of the sum is 541.\
So the reverse digits are 5, 4 and 1.\
Thus putting them into a list, the output becomes [ 5, 4, 1 ]


```python
limit =  int(input())

total = 0
for i in range(1, limit + 1):
    if i % 2 != 0:
        one_string = "1"*i
        total += int(one_string)
    else:
        three_string = "3"*i
        total += int(three_string)


print("Sum of series:", total)

rev_total = str(total)
rev_total = rev_total[-1::-1]

rev_list = list()
for i in range(len(rev_total)):
    rev_list.append(int(rev_total[i]))

print("Reverse sum of series in list:", rev_list)
```

### Question-2:

Write a Python program that takes a string as an input where multiple numbers are separated by hashes(#).

* Your first task is to create a list of numbers and print it.
* Your second task is to create a dictionary that will have the index of the list as key, and for the even indices,
multiplication from start to that index as value, while for the odd indices, summation from start to that index as value. Finally, print the dictionary.

=======================================================================

**Sample Input 1:**

1#2#3#4

**Sample Output 1:**

[1, 2, 3, 4]\
{0: 1, 1: 3, 2: 6, 3: 10}

**Explanation 1:**\
key 0 -> value 1\
key 1 -> value 1+2 = 3\
key 2 -> value 1x2x3 = 6\
key 3 -> value 1+2+3+4 = 10\

=======================================================================

**Sample Input 2:**

5#6#7

**Sample Output 2:**
[5, 6, 7]
{0: 5, 1: 11, 2: 210}

**Explanation 2:**\
key 0 -> value 5\
key 1 -> value 5+6 = 11\
key 2 -> value 5x6x7 = 210

```python
hashed_numbers = input()

num_list = hashed_numbers.split("#")
for i in range(len(num_list)):
    num_list[i] = int(num_list[i])

print(num_list)

new_dict = dict()
for i in range(len(num_list)):
    if i % 2 == 0:
        product = 1
        for j in range(i+1):
            product *= num_list[j]
        new_dict[i] = product
    else:
        summ = 0
        for j in range(i+1):
            summ += num_list[j]
        new_dict[i] = summ

print(new_dict)
```

### Question-3:

Write a python function that takes a positive integer n as an argument. This function will return the number of positive integers less than n that are co-prime to n.\
***<font color=cyan>Co-prime Numbers:</font> Two numbers are coprime to each other if they don't have any common divisor except 1.***

=======================================================================

**Sample Input 1:**

1

**Sample Output 1:**

0

**Reason:** Since 1 is the smallest positive integer, there exists no positive integer that is less than and coprime to 1. 

=======================================================================

**Sample Input 2:**

2

**Sample Output 2:**

1

**Reason:** Only 1 doesn't share any common divisor (except 1 itself) with 2.\
Also, 1 < 2.

=======================================================================

**Sample Input 3:**

6

**Sample Output 3:**

2
 
**Reason:** Namely, the two coprime numbers to 6 are 1 and 5.

=======================================================================

**Sample Input 4:**

36

**Sample Output 4:**

12

**Reason:** The numbers coprime to 36 are 1, 5, 7, 11, 13, 17, 19, 23, 25, 29, 31 and 35 


```python
def co_prime_counter(main_number):
    
    co_prime_numList = list()
    co_prime_count = 0
    for number in range(1, main_number):
        
        counter = 0
        for divisor in range(1, number + 1):
            if number % divisor == 0 and main_number % divisor == 0:
                counter += 1
        
        if counter == 1:
            co_prime_count += 1
            co_prime_numList.append(number)
        
    return co_prime_count, co_prime_numList



co_prime_count, co_prime_list = co_prime_counter(36)
print(f"{co_prime_count} --> {co_prime_list}")
co_prime_count, co_prime_list = co_prime_counter(6)
print(f"{co_prime_count} --> {co_prime_list}")
co_prime_count, co_prime_list = co_prime_counter(15)
print(f"{co_prime_count} --> {co_prime_list}")
```

Take two lists from the user (elements of the lists separated by space of the same size). The first list will contain some names and the second list will contain roll numbers.\

Find the sum of the **ASCII** value of each name and then sort the names in descending order based on the sum. Then assign the lowest number to the name that has the highest **ASCII** Value sum as a roll number.

=======================================================================

**Sample Input 1:**

Alan Jack Nancy Jim\
10 15 20 50

**Sample Output 1:**

Nancy's roil number is: 10\
Alan's roll number is: 15\
Jacks roll number is: 20\
Jim's roll number is: 50

**Explanationl:** Here, the summation of the ASCII value of 
each name are:\
Nancy(505), Alan(380), Jack(377) and Jim(288).\
So, Nancy got the lowest number (10) from the number list as roll number, then Alan got the second lowest (15), followed by Jack(20) and Jim (50).

**Sample Input 2:**

Leo Tim Kim Brad Joey\
55 15 25 19 12

**Sample Output 2:**

Joey's roll number is: 12\
Brad's roll number is: 15\
Tim's roll number is: 19\
Kim' roll number is: 25\
Leo's roll number is: 55


```python
name_list = input()
name_list = name_list.split()

roll_list = input()
roll_list = roll_list.split()
roll_list.sort()

ascii_dict = dict()
ascii_values = list()
for i in range(len(name_list)):
    total = 0
    for j in range(len(name_list[i])):
        total += ord(name_list[i][j])
    ascii_dict[total] = name_list[i]
    ascii_values.append(total)

sorted_ascii_values = sorted(ascii_values ,reverse=True)

for i in range(len(sorted_ascii_values)):
    print(f"{ascii_dict[sorted_ascii_values[i]]}'s roll number is {roll_list[i]}")
    
```

### Question-B:

The coordinates of different persons are given in the following dictionary.

position = {'A': (2, 5), 'B': (1.7). 'C': (9, 5), 'D': (2, 6)} 

You will take 2 inputs for person X's location where the first input specifies person X's current location in the x-axis and the second input specifies person X's current location in the y-axis.\
Your work is to find the person closest to person X by calculating the distance between person X and every other person in the dictionary.\
***Hint:*** The distance between two coordinates is given by the formula (You need to import math):

math.sqrt((x2 - x1)\*\*2 + (y2-y1)\*\*2))

=======================================================================

**Sample Input 1:**

Enter person X's x coordinate: 3\
Enter person X's y coordinate: 6

**Sample Output 1:**

D

**Explanation 1:**\
X is at (3.6).\
The distance between person X and person A: 2\
The distance between person X and person B: 5\
The distance between person X and person C: 37\
The distance between person X and person D: 1

The least distance is between person X and D.\
So, D is selected 

=======================================================================

**Sample Input 2:**

Enter person X's x coordinate: 3\
Enter person X's y coordinate: 5

**Sample Output 2:**

A

```python
import math

position = {'A': (2, 5), 'B': (1, 7), 'C': (9, 5), 'D': (2, 6)}
distance_list = list()

x1 = int(input("Enter person X's x coordinate: "))
y1 = int(input("Enter person X's x coordinate: "))

# Creating List of Distances
for value in position.values():
    distance = math.sqrt((value[0] - x1)**2 + (value[1] - y1)**2)
    distance_list.append(distance)

# Finding Minimum Value of Distance

# minimum = distance_list[0]
# for i in range(len(distance_list)):
#     if minimum >= distance_list[i]:
#         minimum = distance_list[i]
#  ------------- OR -------------
minimum = min(distance_list)

# Checking Co-ordinates
for key, value in position.items():
    if math.sqrt((value[0] - x1)**2 + (value[1] - y1)**2) == minimum:
        print(key)
```

### Question-C:

Write a function name "randValues0to9" that takes a string from the user as an argument and generates a random integer between 0 to 9.\
To generate a random value you need a seed. As your range is 0 to 9, by doing `seed%(9+1)` you will get a random value between 0 to 9.\
To generate the seed. you need to count the number of vowels and consonants in that argument. If the vowels are less in number than the consonants, then your seed will be,
> `seed = no_of_vowels*no_of_consonants+21`

Otherwise. your seed will be,

> `seed = no_of_vowels*no_of_consonants+57`

=======================================================================

**Function Call 1:**\
randValues0to9("H@#ello Pe3ter!!")

**Sample Outputl:**\
Vowels: 4\
Consonants: 6\
Seed: 45\
Random value: 5

**Explanation 1:** In "H@#ello Pe3ter!!": string, there are 4 vowels and 6 consonants.\
So the vowels are less in number than the constants and the seed is,
> `seed = no_of_vowels*no_of_consonants+21`

seed = 45\
As your seed is 45, your random value is 45%10 = 5 

=======================================================================

**Function Call 2:**\
randValues0to9("Ayeeeeeeeeeeeee lmao")

**Sample Output2:**\
Vowels: 16\
Consonants: 3\
Seed: 105\
Random value: 5

```python
def randValues0to9(string):

	vowel_string = "aeiouAEIOU"
	no_of_vowels = 0
	no_of_consonants = 0

	for i in range(len(string)):
		if 97 <= ord(string[i]) <= 122 or 65 <= ord(string[i]) <= 90:
			if string[i] in vowel_string:
				no_of_vowels += 1
			else:
				no_of_consonants += 1

	print("Vowels:", no_of_vowels)
	print("Consonants:", no_of_consonants)

	if no_of_vowels < no_of_consonants:
		seed = no_of_vowels * no_of_consonants + 21
	else:
		seed = no_of_vowels * no_of_consonants + 57

	print("Seed:", seed)

	random_val = seed % (9+1)
	print("Random Value:", random_val)


randValues0to9("Ayeeeeeeeeeeeee lmao")
```
