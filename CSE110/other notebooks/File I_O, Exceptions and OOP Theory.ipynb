#File I/O

##*Reading Only (R)*

###Basic File Reading & Output

```python
#############################################
# Specifying mode of operation (read)
#############################################

file = open('Documents/Python_File_IO_files/hello.txt', mode = "r")
print(file.read())
file.close()

#############################################
print("\n#================================#")
print("#================================#\n")
# Without specifying mode of operation (read)
#############################################

file = open('Documents/Python_File_IO_files/hello.txt')
print(file.read())
file.close()

#############################################
print("\n#================================#")
print("#================================#\n")
# Printing within defined index range
# [.read(i), where i is any int() number]
#############################################

file = open('Documents/Python_File_IO_files/hello.txt')
print(file.read(9))
file.close()
```

###Seperate Line Reading & Output

```python
#######################################################
# .read() function for reading all lines normally
#######################################################

# Loop bring extra newline in between:
file = open('Documents/Python_File_IO_files/hello.txt')
for line in file:
    print(line)

file.close()

#############################################
print("\n#================================#")
print("#================================#\n")
# .readline() function does the same
#############################################

file = open('Documents/Python_File_IO_files/hello.txt')
print(file.readline()[:-1]) # string slicing upto -1 to skip extra "\n"
print(file.readline(), end="") # end argument skipping extra "\n"
print(file.readline())

file.close()

#############################################
print("\n#================================#")
print("#================================#\n")
# .readline function can be used in Loop
#############################################

file = open('Documents/Python_File_IO_files/hello.txt')
for i in range(3):
    print(file.readline())
file.close()

#############################################
print("\n#================================#")
print("#================================#\n")
# Making a List to print Between a desired
# range of lines [ .readlines() function to
# create a list]
#############################################

file = open('Documents/Python_File_IO_files/hello.txt')
lines = file.readlines()
print(lines)
for i in range(len(lines)):
    if i == len(lines) - 1:
        print(lines[i]) # Last line [ no need slicing ]
    else:
        print(lines[i][:-1])
        # string slicing upto -1 to skip extra "\n"
file.close()
 
 # OR
print()

file = open('Documents/Python_File_IO_files/hello.txt')
lines = file.readlines()
for line in lines:
    print(line.rstrip()) # auto-stripping to skip extra "\n"
                         # no need extra conditional argument
file.close()
```

##*Writing Only (W)*

###Overwriting
####*\[ mode="w" \]*

```python
#######################################################
# Overwriting a File [ .write() function ]
#######################################################

file = open('Documents/Python_File_IO_files/hello_overwrite.txt', 'w')
file.write("Removed Old data\n") # Overwritten in Existing Old File
file.write("Edited\n") # Data added in same session
file.close()           #(all will be overwriiten if opened with "w" again)

#######################################################

file = open('Documents/Python_File_IO_files/hello_overwrite_new.txt', 'w')
file.write("Data\n") # New File Created & Data Added
file.write("Data\n") # Data Added
file.close()

# No Output will be seen  :)

#######################################################
# Overwriting with .writelines() function
#######################################################

file = open('Documents/Python_File_IO_files/hello_overwrite_new.txt', 'w')
str1 = "Writing this new line.\n"
str2 = "This is the 2nd line.\n"
str3 = "This is the 3rd line.\n"

List_of_lines = [str1, str2, str3]
file.writelines(List_of_lines) # Read from List of Strings & Overwrite
file.close()

# changing the mode of operation, [for Checking]
# write to read
file = open('Documents/Python_File_IO_files/hello_overwrite_new.txt', 'r')
print(file.read())
file.close()
```

###Appending
####*\[ mode='a' \]*

```python
#######################################################
# Appending to an old File/ Creating a new file and appending
# [ same .write() function ]
#######################################################

file = open('Documents/Python_File_IO_files/hello_overwrite.txt', 'a')
file.write("\n[Added with the Old data]\n") # Only New Data Added
file.write("\n") # Same
file.close()

# Again, Output won't be seen  :)

#######################################################
# Appending with .writelines() function
#######################################################

file = open('Documents/Python_File_IO_files/hello_overwrite.txt', 'a')
str1 = "Adding this new line.\n"
str2 = "This is the 2nd line.\n"
str3 = "This is the 3rd line.\n"

List_of_lines = [str1, str2, str3]
file.writelines(List_of_lines) # Read from List of Strings & Append
file.close()

# changing the mode of operation,
# append to read
file = open('Documents/Python_File_IO_files/hello_overwrite.txt', 'r')
print(file.read())
file.close()
```

##*Read & Write (RW)*

```python
# 1. Reading from old file
# 2. Creating new file with edited data from old file

print(f"Data from main file:\n")
file = open('Documents/Python_File_IO_files/Country List.txt')
print(file.read())
file.close()

file = open('Documents/Python_File_IO_files/Country List.txt')
new_file = open('Documents/Python_File_IO_files/Edited Country List.txt', 'w')

for line in file:
    temp = line.split() # Serial No. And Name seperated & Listed inside temp
    new_file.write(f"{temp[1]}\n") # Only writing Country Names from index 1

file.close()
new_file.close()

new_file = open('Documents/Python_File_IO_files/Edited Country List.txt')
print("\nData added after editing in new_file:\n")
print(new_file.read())
new_file.close()
```

#Exceptions

##*Part 1 (Basic Structure)*

```python
try:
    div_calculation = input()
    x, y = div_calculation.split("/")
except ValueError:
    print("\nINPUT ERROR:\nwExpected only numbers with '/' in between\n")
else:
    try:
        x = int(x)
        y = int(y)
        if y == 0:
            raise ZeroDivisionError("\nIntegers non-divisable by 'zero'\n")
    except ValueError:
        print("\nINPUT ERROR:\nExpected '/' found but numbers absent in both sides\n")
    else:
        print("Calculating.....")
        div = x/y
        if int(div) == div:        
            print(f"\n{int(div)}\n")
        else:
            print(f"\n{div}\n")
        
        print("Calculated without ERRORS")
    finally:
        print("Done", end=". ")

print("Thank you.")
```

#OOP (Object Oriented Programming)

##*Creating New Class & Instances*

```python
class Taxi:
    driver_name = "Shafs"
    on_duty = False
    num_passengers = 2
    cities = ["Dhaka", "Comilla"]

taxi1 = Taxi()
print(taxi1.cities)
print(taxi1.driver_name)
print(Taxi.driver_name)
```

```python
class Taxi:
    def __init__(self, name, duty, num, cities):
        self.driver_name = name
        self.on_duty = duty
        self.num_passengers = num
        self.cities = cities
    
    def set_driver_name(self, name):
        self.driver_name = name
    
    def pickup_passenger(self):
        self.num_passengers += 1
        return self.num_passengers
    
    def dropoff_passenger(self):
        self.num_passengers -= 1
        return self.num_passengers
    
    def set_cities(self, city):
        self.cities = city

    def add_cities(self, cities):
        self.cities.append(cities)

taxi1 = Taxi("Shafs", True, 3, ["Dhaka"])

print(taxi1.cities)
print(taxi1.driver_name)
print(taxi1.on_duty)

print()

taxi1.set_driver_name('Ali')
print(taxi1.driver_name)

print()

taxi1.add_cities('Borishal')
print(taxi1.cities)

print()

taxi1.set_cities(['Comilla'])
print(taxi1.cities)

print()

x = taxi1.dropoff_passenger()
print(x)
```
