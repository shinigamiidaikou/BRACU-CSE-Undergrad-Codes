## Task:

#### Implement the ***Group*** Class that is derived from ***Competition*** class so that the following output is generated.
##### [The group name can be of any number of words.]

```python
class Competition:
    info={}
    groupNo = 0

    def __init__(self,group_name,member):
        self.group_name = group_name
        self.total_member = member
        self.total_fee = 0

    def __str__(self):
        return "Groupname: {}\nMembers: {}".format(self.group_name,self.total_member)


class Group(Competition):
	def __init__(self, group_name, member, wins=0):
		super().__init__(group_name, member)
		Competition.groupNo += 1
		group_initals = ""
		group_name_words = group_name.split()
		for initial in group_name_words:
			group_initals += initial[0]
		self.group_id = str(Competition.groupNo)+group_initals+str(self.total_member)
		self.total_win = wins
		self.total_fee = self.total_member * 100
		self.won_competitions = str()
	
	@classmethod
	def createGroup(cls, name, members, wins):
		return cls(name, members, wins)
	
	def __str__(self):
		if self.total_win == 0:
			return "Group ID: "+self.group_id+"\n"+super().__str__()+"\nTotal Fee: "+str(self.total_fee)+"\nTotal win: 0"
		else:
			return "Group ID: "+self.group_id+"\n"+super().__str__()+"\nTotal Fee: "+str(self.total_fee)+"\nTotal win: "+str(self.total_win)+" on "+', '.join(self.won_competitions)
	
	def winningCompetitions(self, names="Not Applicable"):
		if names != "Not Applicable":
			self.won_competitions = names.split(", ")
		else:
			self.won_competitions = names
		Competition.info[self.group_id] = [self.group_name, self.total_member, self.total_win, self.won_competitions]
	
	def competeAgainst(self, refn):
		if self.total_member != refn.total_member:
			print("Need an equal number of members. The groups can not compete against each other.")
		else:
			print("The members are of equal number. The groups can compete against each other.")


print("Number of groups:",Competition.groupNo)
print("Group Database:",Group.info)
print("#################################")
g1 = Group("Abstract Connoisseurs Group", 6, 3)
g1.winningCompetitions("ACM-ICPC, Technovation Challenge, HackerEarth")
print("------Details of the group------")
print(g1)
print("#################################")
g2 = Group("Boolean Autocrats", 3)
g2.winningCompetitions()
print("------Details of the group------")
print(g2)
print("#################################")
g3 = Group.createGroup("Hex Clan",6, 2)
g3.winningCompetitions("TechGig, Hackathon")
print("------Details of the group------")
print(g3)
print("#################################")
print("Number of groups:",Competition.groupNo)
print("group info:",Competition.info)
print("#################################")
g1.competeAgainst(g2)
g1.competeAgainst(g3)
g3.competeAgainst(g2)
```
