```python
class Calculator:
    def __init__(self):
        print("Let's Calculate!")
        self.value1 = None
        self.value2 = None
        self.result = None
    def add(self):
        self.result = self.value1 + self.value2
    def subtract(self):
        self.result = self.value1 - self.value2
    def multiply(self):
        self.result = self.value1 * self.value2
    def divide(self):
        self.result = self.value1 / self.value2


x = Calculator()
first_val = int(input("Value 1: "))
operator = input("Operator: ")
second_val = int(input("Value 2: "))

x.value1, x.value2 = (first_val, second_val)

if operator == "+": x.add()
elif operator == '-': x.subtract()
elif operator == '*': x.multiply()
elif operator == '/': x.divide()
print("Result:", x.result)
```

```python
class Shape:
    def __init__(self, s_n, v1, v2):
        self.name = s_n
        self.value1 = v1
        self.value2 = v2
        self.area_val = int()
    def area(self):
        if self.name == "Triangle" or self.name == 'Rhombus':
            self.area_val = 0.5 * self.value1 * self.value2
            print('Area:', self.area_val)
        elif self.name == "Square" or self.name == "Rectangle":
            self.area_val = self.value1 * self.value2
            print('Area:', self.area_val)
        else:
            print("Area: Shape unknown")


triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

### Task 6

```python
class Calculator:
    def __init__(self):
        print("Calculator is ready!")
        self.value1 = None
        self.value2 = None
        self.operator = None
        self.result = None
    def calculate(self, val1, val2, op):
        self.value1 = val1
        self.value2 = val2
        self.operator = op
        if self.operator == "+":
            self.result = self.value1 + self.value2
        elif self.operator == '-':
            self.result = self.value1 - self.value2
        elif self.operator == '*':
            self.result = self.value1 * self.value2
        elif self.operator == '/':
            self.result = self.value1 / self.value2
        return self.result
    def showCalculation(self):
        print(self.value1, self.operator, self.value2, '=', self.result)



c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

### Task 7

```python
class Student:
    def __init__(self, n, id, dep, cg):
        self.name = n
        self.ID = id
        self.dept = dep
        self.grade_list = cg
        self.cgpa = float()
    def calculate_CGPA(self):
        sum_total = 0
        for grade in self.grade_list:
            sum_total += (grade * 3)
        self.cgpa = sum_total / (len(self.grade_list) * 3)
    def print_details(self):
        print(f"Name: {self.name}, ID: {self.ID}")
        print("Department:", self.dept)
        print("CGPA:", self.cgpa)
        if self.cgpa > 3.8:
            print("Your academic standing is 'Highest Distiction'")
        elif 3.8 >= self.cgpa > 3.65:
            print("Your academic standing is 'High Distiction'")
        elif 3.65 >= self.cgpa > 3.5:
            print("Your academic standing is 'Distiction'")
        elif 3.5 >= self.cgpa > 2.0:
            print("Your academic standing is 'Satisfactory'")
        else:
            print("Sorry, you cannot graduate")


s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
class Shinobi:
    def __init__(self, n, r):
        self.name = n
        self.rank = r
        self.mission_count = 0
        self.salary = 0
    def changeRank(self, r):
        self.rank = r
    def calSalary(self, missions):
        self.mission_count = missions
        if self.rank == 'Genin':
            self.salary = self.mission_count * 50
        elif self.rank == 'Chunin':
            self.salary = self.mission_count * 100
        else:
            self.salary = self.mission_count * 500
    def printInfo(self):
        print("Name:", self.name)
        print("Rank:", self.rank)
        print("Number of mission:", self.mission_count)
        print("Salary:", self.salary)


naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
class Spotify:
    def __init__(self, s_l):
        self.playlist = s_l
        print("Welcome to Spotify!")
    def playing_number(self, num):
        if num > len(self.playlist):
            return f"{num} number song not found. Your playlist has {len(self.playlist)} songs only."
        else:
            return f"##########################\nPlaying {num} number song for you\nSong name: {self.playlist[num-1]}"
    def add_to_playlist(self, song):
        self.playlist += [song]


user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
