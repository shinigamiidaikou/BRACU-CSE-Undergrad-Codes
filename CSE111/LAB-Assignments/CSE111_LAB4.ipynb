```python
class Batsman:
    def __init__(self, *details):
        self.name = 'New Batsman'
        self.runs = 0
        self.balls = 0
        if len(details) == 3:
            self.name = details[0]
            self.runs = details[1]
            self.balls = details[2]
        else:
            self.runs = details[0]
            self.balls = details[1]
        self.strike_rate = (self.runs/self.balls) * 100
    def setName(self, n):
        self.name = n
    def battingStrikeRate(self):
        return self.strike_rate
    def printCareerStatistics(self):
        print('Name:', self.name)
        print(f'Runs Scored: {self.runs} , Balls Faced: {self.balls}')


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
class StudentDatabase:
    def __init__(self, n, id_num):
        self.name = n
        self.ID = id_num
        self.grades = dict()
    def calculateGPA(self, course_details, session):
        session_details = dict()
        course_list = list()
        totalGPA = 0
        for info in course_details:
            code, grade = list(info.split(": "))
            course_list.append(code)
            totalGPA += float(grade)
        session_details[tuple(course_list)] = totalGPA/len(course_list)
        self.grades[session] = session_details
    def printDetails(self):
        print("Name:", self.name)
        print("ID:", self.ID)
        for session, details in self.grades.items():
            print(f"Courses taken in {session}:")
            for courses, gpa in details.items():
                for course in courses:
                    print(course)
                print("GPA:", round(gpa, 2))


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
