## String

### Task 1
From a given string, print the string in all uppercase if the number of uppercase
letters is more than lowercase letters. Otherwise, if lowercase is greater or equal
to uppercase letters, print all lowercase. The inputs will contain letters (A-Z, a-z)
only.

**Sample Input:**\
HOusE\
ApplE\
BaNaNa

**Sample Output:**\
HOUSE\
apple\
banana


```python
string = input()

upp_count = 0
low_count = 0

for char in string:
	if 97 <= ord(char) <= 122:
		low_count += 1
	else:
		upp_count += 1

if upp_count > low_count:
	print(string.upper())
else:
	print(string.lower())
```

### Task 2
Given a string, print whether it is a number, word or mixed with digit and letters.
If all the characters are numeric values, print NUMBER. If they are all letters,
print WORD. If it is mixed, print MIXED.

```python
alpha_num = input()

count_n = count_a = 0

for char in alpha_num:
	if 97 <= ord(char) <= 122 or 65 <= ord(char) <= 90:
		count_a += 1
	elif 48 <= ord(char) <= 57:
		count_n += 1
	
	if count_n > 0 and count_a > 0:
		print("MIXED")
		break


if count_a > 0 and count_n == 0:
	print("WORD")
elif count_n > 0 and count_a == 0:
	print("NUMBER")
```

### Task 3
In a given string, there will be two uppercase letters in between some lowercase
letters. Print the substring from the first uppercase letter to the last uppercase
letter excluding them. If there are no letters in between them, print the word
BLANK. It is guaranteed that there will be only two uppercase letters in the
string.

```python
mix_str = input()
new_str = str()
i = 0
up_count = 0
while True:
	if 65 <= ord(mix_str[i]) <= 90:
		up_count += 1
	
	if up_count == 1:
		new_str += mix_str[i]
	i += 1
	
	if up_count == 2:
		new_str = new_str[1:]
		if len(new_str) != 0:
			print(new_str)
		else:
			print("BLANK")
		break
```

### Task 4
Create a string from two given strings by concatenating common characters of
the given strings.

```python
str1, str2 = [string for string in input().split(", ")]

common_str = str()
for i in range(len(str1)):
	if str1[i] in str2:
		common_str += str1[i]
for i in range(len(str2)):
	if str2[i] in common_str:
		common_str += str2[i]

if len(common_str) > 0:
	print(common_str)
else:
	print("Nothing in common.")
```

### Task 5
Again, you have lost your USIS password!! You went to the registrar office and requested for a new password. This time, you need to follow some rules to set your password. Otherwise, they won't change it. The rules are:
<ul type=None>
<li>At least one lowercase letter (a-z)</li>
<li>At least one uppercase letter (A-Z)</li>
<li>At least one digit (0-9)</li>
<li>At least one special character ( _ , $ , #, @ )</li>
</ul>

Your task is to find whether a given password follows all those rules. If it breaks any rule, you have to print Lowercase Missing, Uppercase Missing, Digit Missing or Special Missing respective to the missing case. For more than one rule break, print all the rules that were broken (order doesn't matter). If the password is ok, print OK.

**Sample Input:**\
ohMyBR@CU\
ohmybracu\
OhMyBR@CU20

**Sample Output:**\
Digit missing\
Uppercase character missing, Digit missing, Special character missing\
OK

```python
password = input()

low_count = 0
upp_count = 0
num_count = 0
special_count = 0
special_chars = "#_$@"
ok = False

for char in password:
	if 97 <= ord(char) <= 122:
		low_count += 1
	elif 65 <= ord(char) <= 90:
		upp_count += 1
	elif 48 <= ord(char) <= 57:
		num_count += 1
	elif char in special_chars:
		special_count += 1
	
	if all(x>0 for x in [low_count, upp_count, num_count, special_count]):
		print("OK")
		ok = True
		break

if ok is False:
	missing_list = list()
	if upp_count == 0:
		missing_list.append("Uppercase character missing")
	if low_count == 0:
		missing_list.append("Lowercase character missing")
	if num_count == 0:
		missing_list.append("Digit missing")
	if special_count == 0:
		missing_list.append("Special character Missing")
	for i in range(len(missing_list)-1):
		print(missing_list[i], end=", ")
	print(missing_list[-1])
```

## List

### Task 1

Write a python program which prints the frequency of the numbers that were given as input by the user. Stop taking input when you find the string “STOP”. Do not print the frequency of numbers that were not given as input.

```python
numList = list()
num_input = str()

while True:
	
	num_input = input()
	if num_input == "STOP":
		break
	else:
		numList.append(num_input)

key_list = list()
for i in range(len(numList)):
	if numList[i] not in key_list:
		key_list.append(numList[i])

for i in range(len(key_list)):
	counter = numList.count(key_list[i])
	print(f"{key_list[i]} - {counter} times")
```

### Task 2

Write a python program that calculates the sum of N given lists and prints the highest sum and its respective list. Input starts with N and followed by N lists.

```python
n = int(input())
n -= 1

highest_sumList = [int(x) for x in input().split()]
highest_sum = sum(highest_sumList)

for i in range(n):
	list_in = [int(x) for x in input().split()]
	total = sum(list_in)
	if total > highest_sum:
		highest_sum = total
		highest_sumList = list_in

print(f"{highest_sum}\n{highest_sumList}")
```

### Task 3
Let there be N numbers in a list which is said to be a UB Jumper if the absolute values of the difference between the successive elements take on all the values 1 through N − 1. For example, 2 1 4 6 10 is a UB Jumper because the absolute differences between them are 1 3 2 4 which is all numbers from 1 to (5 - 1) or 4.\
Write a python program that takes a number sequence as input and prints whether it is a UB Jumper or Not UB Jumper. Input will stop after getting “STOP” as input. (Number order or absolute difference order doesn’t follow any sequence.)

```python
while True:

	num_list = input()
	if num_list == "STOP":
		break
	else:
		num_list = num_list.split()
		for i in range(len(num_list)):
			num_list[i] = int(num_list[i])
			
		n_list = list(range(1,len(num_list)))

	diff_list = list()
	for j in range(len(n_list)):
		diff = abs(num_list[j] - num_list[j+1])
		if diff in n_list:
			if diff not in diff_list:
				diff_list.append(diff)
			else:
				print("Not UB Jumper")
				break
		else:
			print("Not UB Jumper")
			break

	if len(diff_list) == len(n_list):
		print("UB Jumper")
```

```python
dict1_in = [x for x in input().split(", ")]
dict2_in = [x for x in input().split(", ")]

dict1 = dict()
for i in range(len(dict1_in)):
	temp = dict1_in[i].split(":")
	dict1[temp[0]] = int(temp[1])

dict2 = dict()
for i in range(len(dict2_in)):
	temp = dict2_in[i].split(":")
	dict2[temp[0]] = int(temp[1])

main_dict = dict()
for key in dict1.keys():
	main_dict[key] = dict1[key]
for key in dict2.keys():
	if key not in main_dict.keys():
		main_dict[key] = dict2[key]
	else:
		main_dict[key] += dict2[key]
print(main_dict)

temp = list()
for value in main_dict.values():
	if value not in temp:
		temp.append(value)

sorted_values = tuple(sorted(temp))
print('Values:', sorted_values)
```

```python

def bmi_calculate(height, weight):
	height = height/100
	bmi = weight/(height*height)
	if bmi > 30:
		print(f"Score is {bmi:.1f}. You are Obese")
	elif 30 >= bmi >= 25:
		print(f"Score is {bmi:.1f}. You are Overweight")
	elif 24.9 >= bmi >= 18.5:
		print(f"Score is {bmi:.1f}. You are Normal")
	else:
		print(f"Score is {bmi:.1f}. You are Underweight")


bmi_calculate(175, 96)
bmi_calculate(152, 48)
```

```python
def divisable_numbers_summ(min, max, div):
	div_total = 0
	for i in range(min, max):
		if i % div == 0:
			div_total += i
	
	return div_total


print(divisable_numbers_summ(0, 10, 2))
print(divisable_numbers_summ(3, 16, 3))
```

```python
def replace_domain(email, new_domain, old_domain='sheba.xyz'):
	email_parts = email.split('@')
	if old_domain == new_domain:
		return f'Unchanged: {email}'
	else:
		return f'Changed: {email_parts[0]}@{new_domain}'

print(replace_domain('alice@kaaj.com', 'sheba.xyz', 'kaaj.com'))
print(replace_domain('bob@sheba.xyz','sheba.xyz'))
```

```python
def palindrome_checker(string):
	string = string.replace(' ', '')
	x = 0
	y = len(string) - 1
	ispalindrome = True
	while(x < y):
		if string[x] != string[y]:
			ispalindrome = False
			break
		x += 1
		y -= 1
	if ispalindrome is True:
		return 'Palindrome'
	else:
		return 'Not a palindrome'

print(palindrome_checker('madam'))
print(palindrome_checker('hello'))
print(palindrome_checker('nurses run'))
```

```python
def capitalization_fix(string):
	fixed_string = str()
	if 90 <= ord(string[0]) <= 122:
		fixed_string += chr(ord(string[0]) - 32)
	else:
		fixed_string += string[0]
	i = 1
	while i < len(string):
		if string[i] in '.?!' and i != len(string) - 1:
			fixed_string += string[i] + ' '
			i += 2
			if 90 <= ord(string[i]) <= 122:
				fixed_string += chr(ord(string[i]) - 32)
				i += 1
			else:
				fixed_string += string[i]
				i += 1
		elif string[i-1] == ' ' and string[i] == 'i' and string[i+1] == ' ':
			fixed_string += 'I'
			i += 1
		else:
			fixed_string += string[i]
			i += 1
	return fixed_string


print(capitalization_fix("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog's name? i love my pet very much."))
```
