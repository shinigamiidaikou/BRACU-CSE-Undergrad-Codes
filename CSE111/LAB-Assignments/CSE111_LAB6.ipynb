```python
class Assassin:
	serial = 0
	def __init__(self, name, success):
		self.name = name
		self.success_rate = success
		Assassin.serial += 1
	
	@classmethod
	def failureRate(cls, name, failure):
		return cls(name, 100 - failure)
	
	@classmethod
	def failurePercentage(cls, name, percentage):
		return cls(name, 100 - int(percentage[:-1]))
	
	def printDetails(self):
		print("Name:", self.name)
		print("Success rate:", self.success_rate)
		print("Total number of Assassin:", Assassin.serial)


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
from datetime import date

class Employee:
	def __init__(self, name, work_period):
		self.name = name
		self.workingPeriod = work_period
	
	@classmethod
	def employeeByJoiningYear(cls, name, joiningYear):
		return cls(name, date.today().year - joiningYear)
	
	@staticmethod
	def experienceCheck(years, gender):
		if gender == 'male':
			if years >= 3:
				return "He is experienced"
			else:
				return "He is not experienced"
		else:
			if years >= 3:
				return "She is experienced"
			else:
				return "She is not experienced"


employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
import math

class Cylinder:
	radius = 5
	height = 18

	def __init__(self, rad, hght):
		self.radius = rad
		self.height = hght
		print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
		print(f"Updated: radius={self.radius} and height={self.height}.")
		Cylinder.radius = self.radius
		Cylinder.height = self.height
	
	@staticmethod
	def area(rad, hght):
		area = 2*math.pi*rad*(rad + hght)
		print("Area: ", area)
	
	@staticmethod
	def volume(rad, hght):
		volume = math.pi*(rad**2)*hght
		print("Volume: ", volume)

	@classmethod
	def swap(cls, hght, rad):
		return cls(rad, hght)
	
	@classmethod
	def changeFormat(cls, string):
		return cls(int(string.split('-')[0]), int(string.split('-')[1]))


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class SultansDine:
	total_branch = 0
	total_sell = 0
	branch_details = dict()

	def __init__(self, b_name):
		self.branch = b_name
		self.sell_quantity = 0
		self.branch_sell = 0
		SultansDine.total_branch += 1
	
	def sellQuantity(self, quantity):
		self.sellQuantity = quantity
		if quantity < 10:
			self.branch_sell = quantity * 300
		elif 10 <= quantity < 20:
			self.branch_sell = quantity * 350
		else:
			self.branch_sell = quantity * 400
		SultansDine.total_sell += self.branch_sell
		SultansDine.branch_details[self.branch] = self.branch_sell
	
	def branchInformation(self):
		print("Branch Name:", self.branch)
		print("Branch Sell:", self.branch_sell, "Taka")
	
	@classmethod
	def details(cls):
		print("Total Number of branch(s):", SultansDine.total_branch)
		print("Total Sell:", SultansDine.total_sell, "Taka")
		for name, sell_value in SultansDine.branch_details.items():
			print(f"Branch Name: {name}, Branch Sell: {sell_value} Taka")
			percentage = round((sell_value/cls.total_sell)*100, 2)
			print(f"Branch consists of total sell's: {percentage}%")




SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```

```python
## Must be initialized Firsthand:

def autoDict(dict_refn, v_self, key, value):
    if v_self not in dict_refn.keys():
        dict_refn[v_self]  = dict()
    if key not in dict_refn[v_self].keys():
        if value is None:
            dict_refn[v_self][key] = ["None"]
        else:
            dict_refn[v_self][key] = [value]
    else:
        if value is None:
            dict_refn[v_self][key] += ["None"]
        else:
            dict_refn[v_self][key] += [value]
```
