```python
class house:

	def __init__(self):
		self.window = 4
		self.door = 2
	
	def view(self):
		print(f"{self.window} windows, {self.door} doors")

#============================================================

house1 = house()
house2 = house()

house1.view()
house2.view()

house1.window += 2
house2.door -= 1

house1.view()
house2.view()
print(house2.__dict__)
print(dir(house1))
```

```python
class Cat:
	def __init__(self, color, action):
		self.color = color
		self.action = action
	
	def view(self):
		print(self.color, "cat is", self.action)
	
	def compare(self, cat) : 
		if self.action == cat.action:
			print("Both cats are", self.action)
		else:
			print("They are different")

c1 = Cat("White", 'jumping') 
c2 = Cat("Brown", "sitting")

c1.view() 
c2.view() 

c1.compare(c2)

c2.action = "jumping"

print()
c1.view() 
c2.view()

c1.compare(c2)
```

```python
class calc:
	
	def summation(self, *nums):
		total = 0
		for i in range(len(nums)):
			total += nums[i]
		print(f"Summation is {total}")
	
	def substraction(self, num1, num2):
		print(num1 - num2)
	
	def product(self, nums):
		product = 0
		for i in range(len(nums)):
			product *= nums[i]
		print(f"Product is {product}")
	
	def division(self, num1, num2):
		print(num1 / num2)


c1 = calc()
c1.summation(4, 5)
```

```python
from multipledispatch import dispatch
class prod:

	@dispatch(int, int)
	def product(self, num1, num2):
		product = num1 * num2
		print(f"Product is {product}")

	@dispatch(int, int, int)
	def product(self, num1, num2, num3):
		product = num1 * num2 * num3
		print(f"Product is {product}")

c1 = prod()
c1.product(4, 5)
c1.product(4, 5, 6)
```
