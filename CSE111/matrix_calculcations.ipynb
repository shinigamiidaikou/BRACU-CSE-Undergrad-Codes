### Matrix Input: (Please input Matrices here)

```python
def matrix_input():
    matrix_row = int(input("Rows: "))
    matrix_columns = int(input("Columns: "))
    matrix = list()
    for i in range(matrix_row):
        row = [int(x) for x in input().split()]
        matrix.append(row)
    
    return matrix

x = matrix_input()
y = matrix_input()

print("Matrix 1:")
print_matrix_list(x)
print("Matrix 2:")
print_matrix_list(y)
```

### Matrix Summation

```python
def matrix_summ(matrix_a, matrix_b):
	new_matrix = list()
	
	if len(matrix_a) == len(matrix_b):
		equal_state = True
		for i in range(len(matrix_a)):
			if len(matrix_a[i]) != len(matrix_b[i]):
				equal_state == False
				break
		if equal_state is False:
			return "No of rows not same"
		else:
			row_matrix = list()
			for i in range(len(matrix_a)):
				for j in range(len(matrix_a[i])):
					row_item = matrix_a[i][j] + matrix_b[i][j]
					row_matrix.append(row_item)
				new_matrix.append(row_matrix)
				row_matrix = list()
			return new_matrix
	else:
		return "No of columns not same"


given_matrix_a = [
	[1, 5, 4],
	[-4, 3, 3],
	[-4, 0, 6]
]

given_matrix_b = [
	[2, 3, 4],
	[8, -3, 9],
	[3, 2, 5]
]

x = matrix_summ(given_matrix_a, given_matrix_b)
print_matrix_list(x)
```

### Matrix Substraction

```python
def matrix_substract(matrix_a, matrix_b):
	new_matrix = list()
	
	if len(matrix_a) == len(matrix_b):
		equal_state = True
		for i in range(len(matrix_a)):
			if len(matrix_a[i]) != len(matrix_b[i]):
				equal_state == False
				break
		if equal_state is False:
			return "No of rows not same"
		else:
			row_matrix = list()
			for i in range(len(matrix_a)):
				for j in range(len(matrix_a[i])):
					row_item = matrix_a[i][j] - matrix_b[i][j]
					row_matrix.append(row_item)
				new_matrix.append(row_matrix)
				row_matrix = list()
			return new_matrix
	else:
		return "No of columns not same"


given_matrix_a = [
	[10, 5, 4],
	[-4, 3, 3],
	[-4, 0, 6]
]

given_matrix_b = [
	[2, 3, 4],
	[8, -3, 9],
	[3, 2, 5]
]

x = matrix_substract(given_matrix_a, given_matrix_b)
print_matrix_list(x)
```

### Matrix Multiplication

```python
def matrix_multiply(matrix_a, matrix_b):
    multiplied_matrix = list()
    if len(matrix_a[0]) != len(matrix_b):
        print("Can't be multiplied.")
    else:
        row_list = list()
        for m in range(len(matrix_a)):
            for i in range(len(matrix_b[0])):
                element = 0
                for j in range(len(matrix_b)):
                    element += matrix_a[m][j] * matrix_b[j][i]
                row_list.append(element)
            multiplied_matrix.append(row_list)
            row_list = list()
        return multiplied_matrix


given_matrix_a = [
	[10, 5, 4],
	[-4, 3, 3]
]

given_matrix_b = [
	[2, 3],
	[8, -3],
	[3, 2]
]

x = matrix_multiply(given_matrix_a, given_matrix_b)
print_matrix_list(x)
```

### Matrix Transpose

```python
def transpose_matrix(m):
    t = list()
    for i in range(len(m)):
        if i == 0:
            for j in range(len(m[i])):
                t.append([m[i][j]])
        else:
            for j in range(len(m[i])):
                t[j].append(m[i][j])
    return t


given_matrix_a = [
	[1, 5, 4],
	[-4, 3, 3],
	[-4, 0, 6]
]

x = transpose_matrix(given_matrix_a)
print_matrix_list(x)
```
