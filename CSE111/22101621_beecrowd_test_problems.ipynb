### 1238 - <font color=#436cb5>Combiner</font> - beecrowd

```python
############################################
#=============== ACCEPTED ==================
############################################

def combiner(main_string):
	Left_string, right_string = main_string.split()

	if len(Left_string) <= len(right_string):
		for i in range(len(Left_string)):
			print(Left_string[i], end="")
			print(right_string[i], end="")
		print(right_string[len(Left_string):])
	
	else:
		for i in range(len(right_string)):
			print(Left_string[i], end="")
			print(right_string[i], end="")
		print(Left_string[len(right_string):])


n = int(input())
for i in range(n):
	string = input()
	combiner(string)
```

### 1287 - <font color=#436cb5>Friendly Int Parser</font> - beecrowd

```python
############################################
#=============== ACCEPTED ==================
############################################

def integer_parser(string):
	string = string.replace(" ", "")
	string = string.replace(",", "")

	parsabilty = True
	parsed_number = str()
	if len(string) > 0:
		for i in range(len(string)):
			if string[i] == "O" or string[i] == "o":
					parsed_number += "0"
			elif string[i] == "l":
				parsed_number += "1"
			elif 48 <= ord(string[i]) <= 57:
				parsed_number += string[i]
			else:
				parsabilty = False
				break
	else:
		return "error"

	if parsed_number == "" or not(0 <= int(parsed_number) < 2147483648):
		return "error"
	elif parsabilty is False:
		return "error"
	else:
		return int(parsed_number)

print_list = list()
while True:
	try:
		string = input()
		print_list.append(integer_parser(string))
	except EOFError:
		break

for i in range(len(print_list)):
	print(print_list[i])
```
