### Number Problem:
<font size = 4></B>
Given a strictly increasing sequence of six positive integers such that their sum is 79, and each of them divides the next one, find the sequence.
</B></font>

```python
def calcNumbers(num, muls):
	for i in range(1, len(num)):
		num[i] = num[i-1] * muls[i-1]

num = [1]+[1]*6
```

```python
muls = [1,2,2,2,3,4]
calcNumbers(num,muls)
print(f"{' + '.join(str(_) for _ in num[1:])} = {sum(num[1:])}")
```

```python
# Main Code

def calcNumbers(num, muls):
	## Sets all numbers according to it's previous number times multiplier
	for i in range(1, len(num)):
		num[i] = num[i-1] * muls[i-1]


target = 135 		## Target Score
num = [1]*7			## Number Array
muls = [1]+[2]*5	## Multipliers Array
i = len(muls)-2		## Initially last multiplier
j = len(muls)-1		## Initially 2nd last multiplier
calcNumbers(num, muls)
impossible = False
while True:
	while j < len(muls)-1:		## While & If "j" is not last position,
		i += 1					## ... Keep incrementing i, j
		j += 1
	if sum(num[1:]) == target:	## Base Condition (Summ Matches Total)
		break
	elif sum(num[1:]) < target:	## If smaller, increase "j"-th Multiplier value
		muls[j] += 1
	else:						## If bigger,
		muls[j] = 2				## ... set "j"-th multiplier <= 2
		muls[i] += 1			## ... increment "i"-th multiplier, +1
		calcNumbers(num, muls)
		while sum(num[1:]) > target:			## While & If summ stil bigger,
			muls[i] = 2						## ... reset "i"-th to 2
			if i > 1:						## ... if "i" at greater than 1st position,
				i -= 1						## ... ... back i, j left by 1
				j -= 1
				muls[i] += 1				## ... ... increment "i"-th multiplier, +1
			elif i == 1:					## ... else if "i" at 1st position,
				muls[0] += 1				## ... ... increment "1st number" by 1
				calcNumbers(num, muls)
				if sum(num[1:]) > target:	## ... ... if still bigger/(not possible),
					impossible = True  ##{***} ... ... ... "IT'S IMPOSSIBLE"
					break					## ... ... ... "GET OUT OF THIS LOOP"
			calcNumbers(num, muls)
	if impossible:				## If "IMPOSSIBLE",
		break					## ... "Get out of this one too"
	calcNumbers(num, muls)
	print(f"{' + '.join(str(_) for _ in num[1:])} = {sum(num[1:])}")


print("="*25)
if impossible:
	print("Not Possible!")
else:
	print(f"Answer:\n{' + '.join(str(_) for _ in num[1:])} = {target}")
print("="*25)
```
