### Imports and Functions

```python
from time import *
from random import *

def randomArr(length):
	arr = []
	for i in range(length):
		arr.append(randint(1,length))
	return arr

def selectionSort(arr):
	for i in range(len(arr) - 1):
		min_idx = i
		for j in range(i+1, len(arr)):
			if arr[min_idx] > arr[j]:
				min_idx = j
		arr[i], arr[min_idx] = arr[min_idx], arr[i]
```

### Selection Sort Test:

```python
print("Selection Sort Times:")

arr = randomArr(5000)

tic = perf_counter()
selectionSort(arr)
toc = perf_counter()

print("5000 Elements:  ", toc - tic)

arr = randomArr(50000)

tic = perf_counter()
selectionSort(arr)
toc = perf_counter()

print("50000 Elements:  ", toc - tic)

arr = randomArr(500000)

tic = perf_counter()
selectionSort(arr)
toc = perf_counter()

print("500000 Elements:  ", toc - tic)
```
