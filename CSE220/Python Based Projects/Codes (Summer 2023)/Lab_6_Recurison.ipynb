```python
#a
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)


print(factorial(5))
```

```python
#b
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


for i in range(8):
    print(fibonacci(i), end=" ")
```

```python
#c
def print_array(arr, index):
    if index == len(arr) - 1:
        print(arr[index])
    else:
        print(arr[index], end=", ")
        print_array(arr, index + 1)


arr = [i for i in range(1, 11)]
print_array(arr, 0)
```

```python
#d
def powerN(base, n):
    if n == 0:
        return 1
    else:
        return base * powerN(base, n - 1)


for i in range(1,5):
    print(powerN(3, i))
```

```python
#2.a
def decimal_to_binary(n):
    if n == 0:
        return '0'
    elif n == 1:
        return '1'
    else:
        return decimal_to_binary(n // 2) + str(n % 2)


print(decimal_to_binary(15))
print(decimal_to_binary(27))
print(decimal_to_binary(127))
```

```python
# SinglyNode Class for 2b and 2c
class SinglyNode:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
```

```python
#2b
def recursive_list_sum(node):
    if node is None:
        return 0
    else:
        return node.data + recursive_list_sum(node.next)


i = 1
j = 9
n = SinglyNode(i)
head = n
for i in range(i+1, j+1):
    n.next = SinglyNode(i)
    n = n.next

# created linkedList = [ integers 'i' to 'j' ]; head = integer i
recursive_list_sum(head)
```

```python
#2c
def print_reverse_linked_list(node):
    if node is None:
        return
    else:
        print_reverse_linked_list(node.next)
        print(node.data)


i = 10
j = 50
diff = 10
n = SinglyNode(i)
head = n
for i in range(i+diff, j+1, diff):
    n.next = SinglyNode(i)
    n = n.next

print_reverse_linked_list(head)
```

```python
#3
def hocBuilder(height):
    if height <= 0:
        return 0
    elif height == 1:
        return 8
    else:
        return 5 + hocBuilder(height - 1)


for i in range(4):
    print(hocBuilder(i))
```

```python
#4a
def linePrint(n):
    if n == 1:
        print(1, end="")
    else:
        linePrint(n-1)
        print(" "+str(n), end="")


def print_pattern(n, current=1):
    if current > n:
        return
    else:
        linePrint(current)
        print()
        print_pattern(n, current + 1)


print_pattern(5)
```

```python
#4b
def linePrint(n):
    if n == 1:
        print(1, end="")
    else:
        linePrint(n-1)
        print(" "+str(n), end="")


def print_pattern(n, current=1):
    if current > n:
        return
    else:
        print(" "*(n-current)*2, end="")
        linePrint(current)
        print()
        print_pattern(n, current + 1)


print_pattern(6)
```

```python
#5
import sys
sys.setrecursionlimit(10000)

class FinalQ:
    def printProfit(self, array, idx):
        if idx < len(array):
            profit = self.calcProfit(array[idx])
            print(f"{idx+1}. Investment: {array[idx]}; Profit: {profit:.1f}")
            self.printProfit(array, idx + 1)

    def calcProfit(self, investment):
        if investment <= 25000:
            return 0
        elif 25000 < investment <= 100000:
            return (investment-25000) * 0.045
        else:
            return (investment-100000) * 0.08 + self.calcProfit(100000)


#Tester
array=[25000,80000,99000,100000,250000,350000]
f = FinalQ()
f.printProfit(array,0)
```
