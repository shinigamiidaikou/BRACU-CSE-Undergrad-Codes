```python
# Complete the functions defined in this cell
#============================
# Extra function:
def sizeOfArray(arr):
    size = 0
    while size < len(arr):
        if arr[size] == None:
            break
        size += 1
    return size
#============================


# Test 01: Play Right
def playRight(sequence, beats):
    for i in beats:
        if i==1:
            temp=sequence[len(sequence)-1]
            for j in range(len(sequence)-1,0,-1):
                sequence[j]=sequence[j-1]
            sequence[0]=temp
    return sequence


# Test 02: Discard Cards
def removeElement(arr, size, idx):
    i = idx
    while i < size - 1:
        arr[i] = arr[i+1]
        i += 1
    arr[i] = None


def discardCards(cards, num):
    size = sizeOfArray(cards)
    i = 0
    while i < size:
        if cards[i] == num:
            removeElement(cards, size, i)
            size -= 1
            continue
        i += 1
    return cards


# Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
    merged_lineup = np.array([None]*len(pokemon_1), dtype=object)
    i, j = 0, len(pokemon_1) - 1
    while i < len(pokemon_1):
        if pokemon_1[i] == None or pokemon_2[j] == None:
            if pokemon_1[i] == None and pokemon_2[j] == None:
                merged_lineup[i] = 0
            elif pokemon_1[i] == None:
                merged_lineup[i] = 0 + pokemon_2[j]
            else:
                merged_lineup[i] = pokemon_1[i] + 0
        else:
            merged_lineup[i] = pokemon_1[i] + pokemon_2[j]
        i += 1
        j -= 1

    return merged_lineup
    

# Test 04: Balance your Salami
def balanceSalami(salami):
    for i in range(len(salami)):
        selfTotal, siblingTotal = 0, 0
        for j in range(0, i):
            selfTotal += salami[j]
        for j in range(i, len(salami)):
            siblingTotal += salami[j]
        if selfTotal == siblingTotal:
            return True
            break
    return False


# Test 05: Protecc Salami
def protectSalami(salami):
    max = salami[0]
    for i in range(1, len(salami)):
        if salami[i] > max:
            max = salami[i]
    
    countArray = np.array([0]*(max+1))
    for i in range(len(salami)):
        countArray[salami[i]] += 1
    
    for i in range(len(countArray)):
        if countArray[i] > 1:
            for j in range(i+1, len(countArray)):
                if countArray[i] == countArray[j]:
                    return True
    return False


# Test 05: Odd Even Wave
def waveYourFlag(arr):
    waveArray = np.array([0]*len(arr))
    waveArray[0] = arr[0]
    if arr[0] % 2 == 0:
        evenArray = np.array([0]*(len(arr)//2))
        oddArray = np.array([0]*(len(arr)//2 + 1))
    else:
        evenArray = np.array([0]*(len(arr)//2 + 1))
        oddArray = np.array([0]*(len(arr)//2))
    eSize, oSize = 0, 0
    for i in range(1, len(arr)):
        if arr[i] % 2 == 0:
            evenArray[eSize] = arr[i]
            eSize += 1
        else:
            oddArray[oSize] = arr[i]
            oSize += 1
    eSize, oSize = 0, 0
    for i in range(1, len(arr), 2):
        if arr[0] % 2 == 0:
            if i == len(arr) - 1:
                waveArray[i] = oddArray[oSize]
                break
            else:
                waveArray[i] = oddArray[oSize]
                waveArray[i+1] = evenArray[eSize]
                eSize += 1
                oSize += 1
        else:
            if i == len(arr) - 1:
                waveArray[i] = evenArray[eSize]
                break
            else:
                waveArray[i] = evenArray[eSize]
                waveArray[i+1] = oddArray[oSize]
                eSize += 1
                oSize += 1
    return waveArray
```
